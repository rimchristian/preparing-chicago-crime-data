```python
# this cell can take up to 1 min to run
chicago_full['Datetime'] = pd.to_datetime(chicago_full['Date'], format=date_format)
chicago_full = chicago_full.sort_values('Datetime')
chicago_full = chicago_full.set_index('Datetime')
chicago_full
```

## Separate the Full Dataset by Years

```python
# save the years for every crime
chicago_full["Year"] = chicago_full.index.year
chicago_full["Year"] = chicago_full["Year"].astype(str)
chicago_full["Year"].value_counts()
```

```python
## Dropping unneeded columns to reduce file size
drop_cols = ["X Coordinate","Y Coordinate", "Community Areas","FBI Code",
             "Case Number","Updated On",'Block','Location','IUCR']
```

```python
# save final df
chicago_final = chicago_full.drop(columns=drop_cols).sort_index()#.reset_index()
chicago_final
```

```python
chicago_final.info()
```

```python
chicago_final.memory_usage(deep=True).astype(float)
```

```python
# unique # of year bins
year_bins = chicago_final['Year'].astype(str).unique()
year_bins
```

```python
FINAL_DROP = ['Datetime','Year']#,'Location Description']
```

```python
## set save location 

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
print(f"[i] Saving .csv's to {OUTPUT_FOLDER}")
## loop through years
for year in year_bins:
    
    ## save temp slices of dfs to save.
    temp_df = chicago_final.loc[ year]
    temp_df = temp_df.reset_index(drop=False)
    temp_df = temp_df.drop(columns=FINAL_DROP)

    # save as csv to output folder
    fname_temp = f"{OUTPUT_FOLDER}Chicago-Crime_{year}.csv"#.gz
    temp_df.to_csv(fname_temp,index=False)

    print(f"- Succesfully saved {fname_temp}")
```

## Comparing AM vs. PM Rush Hour

```python
# Extract the hour from the 'Date' column
df['Hour'] = df['Date'].dt.hour
```

```python
# Filter the dataset for crimes during AM rush hour (7 AM - 10 AM)
am_rush_hour_data = df[(df['Hour'] >= 7) & (df['Hour'] <= 10)]

# filter the dataset for crimes during PM rush hour (4 pm - 7 pm)
pm_rush_hour_data = df[(df['Hour'] >= 4) & (df['Hour'] <= 7)]

# Calculate the total number of crimes during AM and PM rush hours
total_am_crimes = len(am_rush_hour_data)
total_pm_crimes = len(pm_rush_hour_data)

# determine which rush hour has more crimes 
if total_am_crimes > total_pm_crimes:
    print('AM rush hour has more crimes.')
elif total_am_crimes < total_pm_crimes:
    print('PM rush hour has more crimes.')

# Get the top 5 most common crimes during AM rush hour
top_5_am_crimes = am_rush_hour_data['Primary Type'].value_counts().head(5)

# Get the top 5 most common crimes during PM rush hour
top_5_pm_crimes = pm_rush_hour_data['Primary Type'].value_counts().head(5)

# Print the top 5 most common crimes during AM and PM rush hours
print("Top 5 most common crimes during AM rush hour:")
print(top_5_am_crimes)
print("\nTop 5 most common crimes during PM rush hour:")
print(top_5_pm_crimes)

# Check if Motor Vehicle Thefts are more common during AM or PM rush hour
motor_vehicle_thefts_am = am_rush_hour_data[am_rush_hour_data['Primary Type'] == 'MOTOR VEHICLE THEFT']
motor_vehicle_thefts_pm = pm_rush_hour_data[pm_rush_hour_data['Primary Type'] == 'MOTOR VEHICLE THEFT']

if len(motor_vehicle_thefts_am) > len(motor_vehicle_thefts_pm):
    print("\nMotor Vehicle Thefts are more common during AM rush hour.")
elif len(motor_vehicle_thefts_am) < len(motor_vehicle_thefts_pm):
    print("\nMotor Vehicle Thefts are more common during PM rush hour.")
else:
    print("\nMotor Vehicle Thefts have an equal occurrence during AM and PM rush hours.")
```

## Comparing Months

```python
# Extract the month from the 'Date' column
df['Month'] = df['Date'].dt.month
```

```python
# calculate the total number of crimes for each month
monthly_crime_counts = df['Month'].value_counts().sort_index()
monthly_crime_counts[:12]
```

```python
# determine the months with the most and least crimes
most_crime_month = monthly_crime_counts.idxmax()
least_crime_month = monthly_crime_counts.idxmin()

print('Month with the most crimes:', most_crime_month)
print('Month with the least crimes:',least_crime_month )
```

```python
# Identify individual crimes that do not follow the overall pattern
monthly_crime_by_type = df.groupby(['Month', 'Primary Type']).size().unstack().fillna(0)
monthly_crime_by_type[:10]
```

```python
# Calculate the difference between the most and least crime months for each crime type
crime_month_diff = monthly_crime_by_type[most_crime_month] - monthly_crime_by_type[least_crime_month]

# Filter the crimes that do not follow the overall pattern
crimes_with_different_patterns = crime_month_diff[crime_month_diff != 0]

# Print the crimes that do not follow the overall pattern
if len(crimes_with_different_patterns) > 0:
    print("\nCrimes with different patterns:")
    print(crimes_with_different_patterns)
else:
    print("\nNo individual crimes deviate from the overall pattern.")
```
